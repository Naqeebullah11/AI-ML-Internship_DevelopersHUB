Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Load the Iris dataset using seaborn

```python
iris = sns.load_dataset("iris")
```

Inspect the dataset

```python
print("Shape of dataset:", iris.shape)
print("\nColumn names:", iris.columns.tolist())
print("\nFirst 5 rows:")
print(iris.head())
```

Summary information

```python
print("\nDataset Info:")
iris.info()

print("\nSummary Statistics:")
print(iris.describe())
```

Scatter plot - Sepal Length vs Sepal Width (colored by species)

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species")
plt.title("Scatter Plot: Sepal Length vs Sepal Width")
plt.show()
```

Histograms of all numeric features

```python
iris.hist(figsize=(10, 8), bins=15, color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Iris Features", fontsize=16)
plt.tight_layout()
plt.show()
```

Box plots to identify outliers

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris, orient="h")
plt.title("Boxplots of Iris Features")
plt.show()
```
